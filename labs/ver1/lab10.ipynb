# Lab 10: Projekt — kompletny system real-time

## Cel

Zbudowanie kompletnego systemu analizy danych w czasie rzeczywistym, łączącego wszystkie poznane technologie.

---

## Wymagania projektu

Projekt realizowany w grupach max. 5-osobowych.

### Architektura systemu

Twój system powinien składać się z następujących komponentów:

1. **Źródło danych** — producent wysyłający dane do Kafki (API, generator, IoT, webscraping)
2. **Apache Kafka** — broker wiadomości
3. **Przetwarzanie** — Spark Structured Streaming lub konsument Pythonowy
4. **Model ML** — klasyfikacja, detekcja anomalii lub regresja
5. **Wyjście** — dashboard, API, zapis do bazy/pliku, alerty

### Minimalne wymagania

- Całość uruchamiana przez `docker compose up`
- Dane przesyłane przez Apache Kafka
- Przynajmniej jedna transformacja strumieniowa (okno czasowe, agregacja)
- Model ML (wytrenowany offline lub online)
- Dokumentacja w README.md
- Prezentacja (5-10 minut)

## Propozycje tematów

Wybierz jeden z poniższych lub zaproponuj własny:

### 1. Detekcja fraudów w transakcjach bankowych
- Producent: generator transakcji (lub dane z Kaggle)
- Model: Isolation Forest / Random Forest
- Alert: powiadomienie o podejrzanej transakcji

### 2. Analiza sentymentu komentarzy w czasie rzeczywistym
- Producent: symulowany strumień recenzji produktów
- Model: klasyfikator sentymentu (scikit-learn / transformers)
- Wyjście: dashboard z trendem sentymentu per produkt

### 3. Monitoring cen produktów (dynamic pricing)
- Producent: scraper/API porównywarki cen
- Przetwarzanie: Spark — średnie kroczace, wykrywanie zmian cen
- Alert: powiadomienie o spadku ceny poniżej progu

### 4. Analiza ruchu na stronie e-commerce
- Producent: generator clickstream (page_view, add_to_cart, purchase)
- Przetwarzanie: segmentacja użytkowników w oknach czasowych
- Model: predykcja konwersji

### 5. Monitoring IoT — czujniki temperatury/wilgotności
- Producent: symulowane dane z czujników
- Przetwarzanie: Spark — detekcja anomalii w oknach czasowych
- Alert: alarm przy przekroczeniu progu
