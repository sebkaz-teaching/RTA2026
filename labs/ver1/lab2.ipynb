## Część 2: Parametry i endpointy

### Zadanie 2.1 — Parametr w URL (path parameter)

Dodaj endpoint `/hello/{name}`, który zwraca `{"message": "Hello {name}!"}`.

````python
%%file app.py
from fastapi import FastAPI

app = FastAPI()

@app.get("/")
def home():
    return {"message": "Witaj w moim API!"}

# TWÓJ KOD
# Endpoint GET /hello/{name}
````

### Zadanie 2.2 — Parametr w URL (query parameter)

Dodaj endpoint `/suma`, który przyjmuje dwa parametry `a` i `b` (float) i zwraca ich sumę:
`GET /suma?a=3.5&b=2.1` → `{"a": 3.5, "b": 2.1, "suma": 5.6}`

````python
# Dodaj endpoint do app.py i przetestuj:

response = requests.get("http://127.0.0.1:8000/suma", params={"a": 3.5, "b": 2.1})
# TWÓJ KOD
````
